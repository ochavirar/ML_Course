## Loading files with the open method

```python
data_3 = open(main_path+"customer-churn-model/Customer Churn Model.txt", "r") # Read mode
```

```python
cols = data_3.readline().strip().split(",")
n_cols = len(cols)
```

```python
# Scirpt to count the number of columns
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in data_3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter+=1

print ("%d rows and %d columns" % (counter, n_cols))
```

```python
df_3 = pd.DataFrame(main_dict)
```

## Writing and reading files

```python
infile = main_path+"customer-churn-model/Customer Churn Model.txt"
out_file = main_path+"customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile) as infile_1:
    with open(out_file, "w") as out_file_1:
        for line in infile_1:
            fields = line.strip().split(",")
            out_file_1.write("\t".join(fields))
            out_file_1.write("\n")
```
